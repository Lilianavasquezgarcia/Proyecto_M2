### PARTE 1: Modelo de Contactos - Programación orientada a objetos (POO)

En esta sección desarrollamos la base de nuestro proyecto definiendo como se estructuran los datos y se crea el class Contacto que agrupa todas las variables que definen a un contacto. Además, en esta sección se usa __int__ como contructor Y self. que nis permitirá mantener una interfaz pública de las variables. Además se usó principios de encapsulamiento con @property para tener una interfaz controlada y gestionar el código de manera interna, al igual que self._

```python
### PARTE 1: Modelo de Contactos - Programación orientada a objetos (POO)

class Contacto:
    def __init__(self, nombre, telefono, email, direccion):
        self.nombre = nombre
        self.telefono = telefono
        self.email = email
        self.direccion = direccion

# Definición de getters de datos y aplicación de rpincipios de encapsulación

    @property
    def nombre(self):
        return self._nombre

    @property
    def telefono(self):
        return self._telefono

    @property
    def email(self):
        return self._email

    @property
    def direccion(self):
        return self._direccion

# Definición de edición de datos

    @nombre.setter
    def nombre(self, nombre):
        self._nombre = nombre

    @telefono.setter
    def telefono(self, telefono):
        self._telefono = telefono

    @email.setter
    def email(self, email):
        self._email = email

    @direccion.setter
    def direccion(self, direccion):
        self._direccion = direccion

    def __str__(self):
        return f"Nombre: {self.nombre}, Teléfono: {self.telefono}, Email: {self.email}, Dirección: {self.direccion}"

    def a_diccionario(self):
        return {
            "nombre": self._nombre,
            "telefono": self._telefono,
            "email": self._email,
            "direccion": self._direccion
        }

```

### PARTE 2: GESTION DE CONTACTOS

En esta sección se desarrolló el almacenamiento de contactos a través de listas y diccionarios. Nuevamente se usa la programación orientada a objetos con la creación de una clase que es el Gestor de Contactos con listas y diccionarios. Se desarrolló los códigos para agregar, modificar (cambiar datos o eliminar) y buscar contactos. Se usó bucles for, se agregó condicionales con if en caso de no existencias, del para eliminar contacto del diccionario, se manejó errores con raise Value Error y se detallo mensajes indicativos al ususario para cada acción.

```python
### PARTE 2: GESTION DE CONTACTOS

# Almacenamiento de contactos a través de listas y diccionarios
class GestorDeContactos:
    def __init__(self):
        self.lista_de_contactos = []
        self.diccionario_de_contactos = {}

    def agregar_contacto(self, nombre, telefono, email, direccion):
        if nombre in self.diccionario_de_contactos:
            raise ValueError(f"El contacto '{nombre}' ya existe")

        for contacto in self.lista_de_contactos:
            if contacto.telefono == telefono:
                raise ValueError(f"El número {telefono} ya se encuentra asociado a otro contacto")


        # Creación y almacenamiento
        nuevo_contacto = Contacto(nombre, telefono, email, direccion)
        self.lista_de_contactos.append(nuevo_contacto)
        self.diccionario_de_contactos[nombre] = nuevo_contacto.a_diccionario()
        print(f"***Contacto {nombre} agregado con éxito.***")


        # Actualizar contacto
    def editar_contacto(self, nombre_actual, nuevo_telefono=None, nuevo_email=None, nueva_direccion=None):
        contacto_encontrado = None
        for c in self.lista_de_contactos:
            if c.nombre == nombre_actual:
                contacto_encontrado = c
                break
        if not contacto_encontrado:
            print(f"Error: El contacto '{nombre_actual}' no existe.")
            return

        if nuevo_telefono:
            for c in self.lista_de_contactos:
                if c.telefono == nuevo_telefono and c.nombre != nombre_actual:
                    print(f"Error: El teléfono {nuevo_telefono} ya lo tiene otro contacto.")
                    return
            contacto_encontrado.telefono = nuevo_telefono

        if nuevo_email:
            contacto_encontrado.email = nuevo_email

        if nueva_direccion:
            contacto_encontrado.direccion = nueva_direccion

        self.diccionario_de_contactos[nombre_actual] = contacto_encontrado.a_diccionario()
        print("\n")
        print(f"*** Contacto '{nombre_actual}' actualizado con éxito. ***")
        print("\n")
        print("Los nuevos datos registrados son:")
        print(f"  > Nombre:    {contacto_encontrado.nombre}")
        print(f"  > Teléfono:  {contacto_encontrado.telefono}")
        print(f"  > Email:     {contacto_encontrado.email}")
        print(f"  > Dirección: {contacto_encontrado.direccion}")

    def eliminar_contacto(self, nombre):
        """Elimina un contacto por su nombre: """
        if nombre not in self.diccionario_de_contactos:
            print(f"Error: No se puede eliminar. El contacto '{nombre}' no existe.")
            return

        self.lista_de_contactos = [c for c in self.lista_de_contactos if c.nombre != nombre]

        del self.diccionario_de_contactos[nombre]

        print("\n")
        print(f"*** Contacto '{nombre}' eliminado con éxito de la agenda. ***")
        print("\n")


        # Busquedas de contactos
    def buscar_por_contacto(self, nombre):
        if nombre not in self.diccionario_de_contactos:
            raise ValueError("Este contacto no existe")
        return self.diccionario_de_contactos[nombre]

    def buscar_por_telefono(self, telefono):
        for contacto in self.lista_de_contactos:
            if contacto.telefono == telefono:
                return contacto
        raise ValueError("Este número no se encuentra registrado")
```
